# Amenities one-hot encoding

```python
# Convert string representation of list to actual Python list
df['amenities'] = df['amenities'].apply(lambda x: x.strip("[]").replace('"', '').split(', '))

# Combine all lists from all rows into a single list
all_list = [item for sublist in df['amenities'] for item in sublist]

# Category Mapping
category_mapping = {
    "Has_Wifi": ["wifi"],
    "TV_Streaming": ["tv", "hdtv", "netflix", "amazon prime video", "hulu", "roku", "apple tv", "fire tv", "chromecast"],
    "Shampoo": ["shampoo"],
    "Conditioner": ["conditioner"],
    "Body_Soap": ["body soap", "body wash"],
    "Outdoor_Recreation": ["pool", "beach", "bowling alley", "hockey rink", "mini golf", "backyard", "outdoor pool", "lake", "ocean view"],
    "Parking_Free": ["free driveway parking", "free parking", "free street parking", "free residential garage"],
    "Parking_Paid": ["paid parking", "paid valet", "paid parking garage", "paid parking lot"],
    "Kitchen_Gas_Stove": ["gas stove"],
    "Kitchen_Electric_Stove": ["electric stove"],
    "Kitchen_Induction_Stove": ["induction stove"],
    "Kitchen_Refrigerator": ["refrigerator"],
    "Kitchen_Oven": ["oven", "double oven", "single oven", "convection oven", "toaster oven"],
    "Sound_System": ["sound system", "bluetooth speaker", "bose", "jbl", "google home", "alexa", "sonos", "marshall"],
    "Heating_Cooling": ["air conditioning", "radiant heating", "central heating", "fan", "central air conditioning"],
    "Safety_Security": ["smoke alarm", "carbon monoxide alarm", "security camera"]
}

# Initialize columns for each category
for category in category_mapping.keys():
    df[category] = 0

# Fill columns based on keyword matching
for i, row in df.iterrows():
    amenities_list = row["amenities"]
    
    for amenity in amenities_list:
        amenity_lower = amenity.lower()
        for category, keywords in category_mapping.items():
            if any(keyword in amenity_lower for keyword in keywords):
                df.at[i, category] = 1

# Save to CSV
df.to_csv("group_2_cleaned_data.csv", index=False)
df
```
